```python
import requests
from bs4 import BeautifulSoup

# URL of the TIOBE Index
url = "https://www.tiobe.com/tiobe-index/"

# Send an HTTP request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, "html.parser")

# Find the programming languages and their ratings
languages = soup.select(".table-striped td:nth-child(1)")
ratings = soup.select(".table-striped td:nth-child(4)")

# Extract data from the HTML elements
language_names = [language.text.strip() for language in languages]
language_ratings = [(rating.text.strip()) for rating in ratings]

# Print the extracted data
for name, rating in zip(language_names, language_ratings):
    print(f"{name}: {rating}")

# Save data for visualization
data = list(zip(language_names, language_ratings))

import matplotlib.pyplot as plt

# Sort the data by popularity
sorted_data = sorted(data, key=lambda x: x[1], reverse=True)

# Extract names and ratings for visualization
names, ratings = zip(*sorted_data)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(names, ratings, color='skyblue')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Top Programming Languages on TIOBE Index')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the chart
plt.show()


```
